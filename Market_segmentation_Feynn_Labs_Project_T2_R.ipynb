```python
yearly_sales_total = parivahar_dashboard_data.groupby("Year")["Total"].sum()
penetration = (yearly_sales_ev_india / yearly_sales_total) * 100

plt.figure(figsize=(12, 17))
bars = plt.bar(yearly_sales_ev_india.index, yearly_sales_ev_india.values, color='blue', label='EV Sales')

for bar, pen in zip(bars, penetration):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 2000,
        f'{pen:.2f}%',
        ha='center',
        va='bottom',
        fontsize=12
    )

plt.xlabel('Year', fontsize=18)
plt.ylabel('EV Sales (In Millions)', fontsize=18)
plt.title('EV Penetration Aaginst All Vehicle Sales In India', fontsize=18)
plt.xticks(yearly_sales_ev_india.index, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
features = ev_data_only[['State', 'Vehicle Class', 'Vehicle Category', 'Total']]
features_encoded = pd.get_dummies(features, columns=['State', 'Vehicle Class', 'Vehicle Category'])
features_encoded.head(6)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
s_features = ss.fit_transform(features_encoded)
s_features
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, silhouette_score

pca = PCA(n_components=2)
pca_features = pca.fit_transform(s_features)
print(pca_features)
```

```python
exp_variance = pca.explained_variance_ratio_
print("Explained Variance for each component:", exp_variance)

mse = mean_squared_error(s_features, pca.inverse_transform(pca_features))
print("Mean Squared Error after PCA:", mse)
```

```python
num_clusters = range(1, 11)

wcss = []
for k in num_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(pca_features)
    wcss.append(kmeans.inertia_)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x=num_clusters, y=wcss, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal Number of Clusters', fontsize=16, fontweight='bold')
plt.xlabel('Number of Clusters', fontsize=14)
plt.ylabel('Sum of Squares (WCSS)', fontsize=14)
plt.xticks(num_clusters)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(pca_features)

ev_data_only['Cluster'] = clusters
sil_score = silhouette_score(pca_features, clusters)
print("Silhouette Score for clustering:", sil_score)
```

```python
pca_data = pd.DataFrame({
    'PC1': pca_features[:, 0],
    'PC2': pca_features[:, 1],
    'Cluster': clusters
})

sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', palette='viridis', data=pca_data, s=100, edgecolor='w')
plt.title('Customer Segmentation using PCA and KMeans', fontsize=20)
plt.xlabel('PCA 1', fontsize=16)
plt.ylabel('PCA 2', fontsize=16)
plt.show()
```

```python
cluster_analysis = ev_data_only.groupby(['Cluster', 'State', 'Vehicle Category']).size()
cluster_analysis = cluster_analysis.reset_index(name='Count')
cluster_analysis
```

```python
dominant_states = []
dominant_vehicle_categories = []
cluster_labels = []

for cluster in ev_data_only['Cluster'].unique():
    print(f"\nCluster {cluster}:")

    cluster_data = cluster_analysis[cluster_analysis['Cluster'] == cluster]
    cluster_dominant_state = cluster_data.groupby('State')['Count'].sum().idxmax()
    cluster_dominant_vehicle_category = cluster_data.groupby('Vehicle Category')['Count'].sum().idxmax()

    dominant_states.append(cluster_dominant_state)
    dominant_vehicle_categories.append(cluster_dominant_vehicle_category)
    cluster_labels.append(f"Cluster {cluster}")

    print(f"Dominant State: {cluster_dominant_state}")
    print(f"Dominant Vehicle Category: {cluster_dominant_vehicle_category}")
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Bar(
    x=cluster_labels,
    y=[cluster_analysis[(cluster_analysis['Cluster'] == cluster) & (cluster_analysis['State'] == state)]['Count'].sum()
       for cluster, state in zip(ev_data_only['Cluster'].unique(), dominant_states)],
    name='Dominant State',
    text=dominant_states,
    textposition='auto'
))

fig.add_trace(go.Bar(
    x=cluster_labels,
    y=[cluster_analysis[(cluster_analysis['Cluster'] == cluster) & (cluster_analysis['Vehicle Category'] == vehicle)]['Count'].sum()
       for cluster, vehicle in zip(ev_data_only['Cluster'].unique(), dominant_vehicle_categories)],
    name='Dominant Vehicle Category',
    text=dominant_vehicle_categories,
    textposition='auto'
))

fig.update_layout(
    barmode='group',
    width=1200,
    height=700,
    title=dict(
        text='Dominant State and Vehicle Category per Cluster',
        font=dict(size=24)
    ),
    xaxis=dict(
        title=dict(
            text='Cluster',
            font=dict(size=22)
        ),
        tickfont=dict(size=14)
    ),
    yaxis=dict(
        title=dict(
            text='Count',
            font=dict(size=22)
        ),
        tickfont=dict(size=14)
    ),
    legend=dict(
        font=dict(size=20)
    )
)

fig.show()
```
